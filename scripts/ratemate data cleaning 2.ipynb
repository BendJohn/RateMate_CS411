```python
df['link_name'] = df['link_name'].astype(str)
df['name_of_course'] = df['name_of_course'].astype(str)
```

```python
df_dict = {}
for major,course in df_as_array:
    course = course.replace(u"\xa0", u" ")
    if major in df_dict:
        df_dict[major].append(course)
    else:
        df_dict[major] = [course]
```

```python
import pprint

# Prints the nicely formatted dictionary
pprint.pprint(df_dict)

```

```python
import csv

with open('grad_requirements.csv', 'w') as csv_file:  
    writer = csv.writer(csv_file)
    for key, value in df_dict.items():
       writer.writerow([key, value])
```
